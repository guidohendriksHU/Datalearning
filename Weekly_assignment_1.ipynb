in steps maak ik gebruik van sep=";" omdat anders de format niet goed gescheiden wordt.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

steps = pd.read_csv("steps.csv", sep=";")
survey = pd.read_csv("survey.csv")
```

```python
df = pd.concat([survey, steps], axis=1)
df.head()
```

```python
df.describe()
```

in bovenstaatde df staan vreemde values in weigtht. zowel 6kilo als 700kilo is niet mogelijk en deze rijen kunnen dus niet worden gebruikt.

```python
df.sort_values("height")
```

in bovenstaande df heeft 1 row geen "height" dit levert problemen op in feature engenering waar zowel weight als height gebruikt worden om BMI te berekenenen. deze rij kan ik niet gebruiken

```python
df["mean"] = df.iloc[:, 9:338].mean(axis=1)
df.head(10)
```

Om te checken of step kolomen vreemde gegevens bevaten moet eerst de mean van deze kolomen worden berekent. Rijen met een hele hoge mean bevatten onmogelijke values. let op ook met een gewone mean kunnen er vreemde values tussen zitten, dit was dan alleen niet genoeg om de mean te beinvloeden. hoge mean rijen worden niet gebruikt

```python
df.sort_values("mean")
```

door de mean values te sorteren kan ik makkelijk zien of hoge mean values aanwezig zijn. die zijn er. mean values boven 20 000 zijn verdacht. om 20 000 stappen te kunnen zetten moet je 5uur lopen. over bijna een jaar zijn de stappen gemeten. mensen hebben niet de tijd om 20uur te lopen per week een jaar lang. deze worden niet gebruikt.

Nu alle vreemde values zijn opgespoort kunnen alle rijen met vreemde values worden geexclude uit de dataset

```python
df_subset_weight = df[(df["weight"] < 200) & (df["weight"] > 40)]
df_subset_mean_steps = df_subset_weight[df_subset_weight["mean"] < 20000]
df_subset_height = df_subset_mean_steps.dropna(subset=["height"])
```

1e rij maakt subset weight, 2e rij code maakt subset mean, 3e rij maakt subset waar de nan values uit height niet worden aangeroepen.

```python
pd.set_option("display.max_columns", None)
df_subset_height
```

pd.set_option geeft mij de mogelijkheid om alle kolomen handmatig te inspecteren in plaats van slechts enkele door de panda's preset te veranderen.

uiteindelijk zijn met cleaning slechts 13 rijen gedropt, de dataset is nu aanzienlijk beter te gebruiken.
